```python
import pandas as pd
from google.colab import files

# 📌 Subir archivo desde el equipo
uploaded = files.upload()
filename = next(iter(uploaded))

# Leer archivo CSV
df = pd.read_csv(filename)

# Reemplazar cadenas vacías por NaN y convertir columna de fecha a datetime
df['Fecha'] = df['Fecha'].replace('', pd.NA)
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Detectar fechas vacías o inválidas
fechas_nulas = df[df['Fecha'].isna()]
print("🔴 Filas con fecha vacía o inválida:")
print(fechas_nulas)
print(f"Total de fechas vacías o inválidas: {fechas_nulas.shape[0]}")

# Eliminar fechas nulas y ordenar por fecha
df_sin_nulos = df.dropna(subset=['Fecha']).sort_values('Fecha')

# Crear un rango continuo de fechas
rango_completo = pd.date_range(start=df_sin_nulos['Fecha'].min(),
                               end=df_sin_nulos['Fecha'].max(),
                               freq='D')

# Detectar fechas faltantes en la secuencia
fechas_existentes = pd.to_datetime(df_sin_nulos['Fecha'].unique())
fechas_faltantes = rango_completo.difference(fechas_existentes)

print("\n🟡 Fechas faltantes en la secuencia temporal:")
print(fechas_faltantes)
print(f"Total de fechas faltantes en la secuencia: {len(fechas_faltantes)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir archivo
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

# Convertir fecha
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Normalizar columna Tipo
df['Tipo'] = df['Tipo'].astype(str).str.upper().str.strip()

# Tipos deseados
tipos_recurso = ['HIDRAULICA', 'TERMICA', 'SOLAR', 'EOLICA']

# Filtrar por tipos válidos
df = df[df['Tipo'].isin(tipos_recurso)].dropna(subset=['Fecha'])

# Filtrar por fecha específica
fecha_objetivo = pd.to_datetime('2024-02-17').date()
df_dia = df[df['Fecha'].dt.date == fecha_objetivo].copy()

# Validar si hay datos
if df_dia.empty:
    print(f"No hay registros para la fecha {fecha_objetivo}.")
else:
    # Colores personalizados
    colores_tipo = {
        'HIDRAULICA': '#1f77b4',
        'TERMICA': '#ff7f0e',
        'SOLAR': '#f1c40f',
        'EOLICA': '#2ca02c'
    }

    # --- Gráfico 1: barras por tipo ---
    energia_por_tipo = df_dia.groupby('Tipo')['Estimacion_kWh'].sum().reindex(tipos_recurso).fillna(0)

    plt.figure(figsize=(8, 5))
    energia_por_tipo.plot(kind='bar', color=[colores_tipo[t] for t in energia_por_tipo.index], edgecolor='black')
    plt.title(f"Energía estimada por tipo de generación - {fecha_objetivo}")
    plt.xlabel("Tipo de generación")
    plt.ylabel("Energía estimada (kWh)")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # --- Gráfico 2: total combinado ---
    plt.figure(figsize=(5, 4))
    plt.bar(['Total'], [energia_por_tipo.sum()], color='mediumseagreen')
    plt.title(f"Total energía estimada - {fecha_objetivo}")
    plt.ylabel("Energía estimada (kWh)")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # --- Gráfico 3: barras apiladas por hora ---
    df_dia['Hora'] = df_dia['Fecha'].dt.hour
    pivot_hora = df_dia.pivot_table(index='Hora', columns='Tipo', values='Estimacion_kWh', aggfunc='sum').fillna(0)
    pivot_hora = pivot_hora.reindex(columns=tipos_recurso, fill_value=0)

    pivot_hora.plot(kind='bar', stacked=True, figsize=(12, 6), color=[colores_tipo[t] for t in tipos_recurso])
    plt.title(f"Participación horaria por tipo de generación - {fecha_objetivo}")
    plt.xlabel("Hora del día")
    plt.ylabel("Energía estimada (kWh)")
    plt.legend(title='Tipo de generación')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir archivo
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

# Convertir columna 'Fecha' a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Normalizar y filtrar tipos de interés
df['Tipo'] = df['Tipo'].astype(str).str.upper().str.strip()
tipos_recurso = ['HIDRAULICA', 'TERMICA', 'SOLAR', 'EOLICA']
df = df[df['Tipo'].isin(tipos_recurso)].dropna(subset=['Fecha'])

# Definir rango semanal: lunes a domingo que incluye 2024-02-12
fecha_inicio = pd.to_datetime('2024-02-12')
fecha_fin = fecha_inicio + pd.Timedelta(days=6)
df_semana = df[(df['Fecha'].dt.date >= fecha_inicio.date()) & (df['Fecha'].dt.date <= fecha_fin.date())].copy()

# Verificar si hay datos
if df_semana.empty:
    print(f"No hay registros entre {fecha_inicio.date()} y {fecha_fin.date()}")
else:
    # Colores diferenciados por tipo
    colores_tipo = {
        'HIDRAULICA': '#1f77b4',
        'TERMICA': '#ff7f0e',
        'SOLAR': '#f1c40f',
        'EOLICA': '#2ca02c'
    }

    # --- Gráfico 1: barras por tipo en la semana ---
    energia_por_tipo = df_semana.groupby('Tipo')['Estimacion_kWh'].sum().reindex(tipos_recurso).fillna(0)

    plt.figure(figsize=(8, 5))
    energia_por_tipo.plot(kind='bar', color=[colores_tipo[t] for t in energia_por_tipo.index], edgecolor='black')
    plt.title(f"Energía estimada por tipo de generación - Semana {fecha_inicio.date()} a {fecha_fin.date()}")
    plt.xlabel("Tipo de generación")
    plt.ylabel("Energía estimada (kWh)")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # --- Gráfico 2: total combinado en la semana ---
    plt.figure(figsize=(5, 4))
    plt.bar(['Total'], [energia_por_tipo.sum()], color='mediumseagreen')
    plt.title(f"Total energía estimada - Semana {fecha_inicio.date()} a {fecha_fin.date()}")
    plt.ylabel("Energía estimada (kWh)")
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # --- Gráfico 3: barras apiladas por día ---
    df_semana['Día'] = df_semana['Fecha'].dt.date
    pivot_dia = df_semana.pivot_table(index='Día', columns='Tipo', values='Estimacion_kWh', aggfunc='sum').fillna(0)
    pivot_dia = pivot_dia.reindex(columns=tipos_recurso, fill_value=0)

    pivot_dia.plot(kind='bar', stacked=True, figsize=(12, 6), color=[colores_tipo[t] for t in tipos_recurso])
    plt.title(f"Participación diaria por tipo de generación - Semana {fecha_inicio.date()} a {fecha_fin.date()}")
    plt.xlabel("Día")
    plt.ylabel("Energía estimada (kWh)")
    plt.legend(title='Tipo de generación')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
